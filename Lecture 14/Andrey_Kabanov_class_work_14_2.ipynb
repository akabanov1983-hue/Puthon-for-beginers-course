```python
import logging
```

```python
my_logger = logging.getLogger()
```

```python
log_handler = logging.FileHandler('my_logs.log', "w", encoding="utf-8")
log_handler.setLevel(logging.ERROR)
my_logger.addHandler(log_handler)
```

```python
try:
    1/0
except ZeroDivisionError as e:
    my_logger.error(f'{e.args} - события уровня error')
except:
    my_logger.critical(f'событие уровня crit')
```
