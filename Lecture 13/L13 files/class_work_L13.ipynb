```python
class User:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"user: {self.name} {self.age}"

    def __repr__(self):
        return f"user: {self.name} {self.age}"


```

```python
d1 = User('D', 1)
k1 = User('K', 2)
```

```python
import pickle

with open('data_picle.txt', 'wb') as fs:
    pickle.dump(d1, fs)
    pickle.dump(k1, fs)

with open('data_picle.txt', 'rb') as fs:
    new_K1 = pickle.load(fs)

new_K1
```
